```python
with open('input.txt') as f:
    input = [line.split()[0] for line in f.readlines()]
```

## PART 1

```python
text = []

len_input = len(input)
width_input = len(input[0])

for input_list in [input, [s[::-1] for s in input[::-1]], [s[::-1] for s in input], [s for s in input[::-1]]]:
    for i in range(0,len_input):
        temp = [element[:i+1] for element in input_list[len_input-i-1:]]
        str = ''
        for j,element in enumerate(temp):
            str+=element[j]
        text.append(str)
        
text.pop(-1) 
text.pop(len_input-1)   

[text.append(item) for item in input]
[text.append(''.join(x)) for x in zip(*input)]

res = 0
for element in text:
    res+=element.count('XMAS')
    res+=element[::-1].count('XMAS')
    
print(res)
    
```

## PART 2

```python
res = 0
for i in range(1,len_input-1):
    for j in range(1,width_input-1):
        temp = [element[j-1:j+2] for element in input[i-1:i+2]] 
        if ([''.join(element[j]) for j,element in enumerate(temp)] in (['M', 'A', 'S'],['S', 'A', 'M'])):
            if ([''.join(element[j]) for j,element in enumerate([s[::-1] for s in temp])] == [''.join(element[j]) for j,element in enumerate(temp)])\
                or ([''.join(element[j]) for j,element in enumerate([s for s in temp[::-1]])] == [''.join(element[j]) for j,element in enumerate(temp)]):
                res+=1
            
print(res)
```
