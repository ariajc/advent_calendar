```python
with open('input.txt') as f:
    lines = [line.strip() for line in f.readlines()]
```

## PART 1

```python
visited_pos = {[(i, line.find('^')) for i, line in enumerate(lines) if line.find('^') != -1][0]:1}
pos = [(i, line.find('^')) for i, line in enumerate(lines) if line.find('^') != -1][0]
dir = (-1,0)

while (0 <= pos[0]+dir[0] < len(lines)) and (0 <= pos[1]+dir[1] <= len(lines[0])):

    if lines[pos[0]+dir[0]][pos[1]+dir[1]] != '#':
        visited_pos[(pos[0]+dir[0],pos[1]+dir[1])] = dir
        pos = (pos[0]+dir[0],pos[1]+dir[1])
        # lines[pos[0]] = lines[pos[0]][:pos[1]] + 'X' + lines[pos[0]][pos[1] + 1:]
    else:
        dir = (dir[1], -dir[0])
        visited_pos[pos] = dir
        # lines[pos[0]] = lines[pos[0]][:pos[1]] + 'X' + lines[pos[0]][pos[1] + 1:]

visited_pos[pos] = dir
print(len(visited_pos))
```

```python
visited_pos = {(i, line.find('^')) for i, line in enumerate(lines) if line.find('^') != -1}
pos = next(iter(visited_pos))
dir = (-1,0)

while (0 <= pos[0]+dir[0] < len(lines)) and (0 <= pos[1]+dir[1] <= len(lines[0])):

    if lines[pos[0]+dir[0]][pos[1]+dir[1]] != '#':
        pos = (pos[0]+dir[0],pos[1]+dir[1])
        visited_pos.add(pos)
        
        # lines[pos[0]] = lines[pos[0]][:pos[1]] + 'X' + lines[pos[0]][pos[1] + 1:]
    else:
        dir = (dir[1], -dir[0])
        visited_pos.add(pos)
        # lines[pos[0]] = lines[pos[0]][:pos[1]] + 'X' + lines[pos[0]][pos[1] + 1:]

visited_pos.add((pos))
print(len(visited_pos))
```
