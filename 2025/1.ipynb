# READ INPUT

```python
with open('input.txt') as f:
    lines = [line.rstrip() for line in f]
```

# PART 1

```python
current_number = 50
count = 0

for line in lines:
    
    step_dir  = -1 if 'L' in line else 1
    step_range = int(line.strip('L').strip('R'))
    
    for _ in range(step_range):
        current_number = (current_number + step_dir) % 100
    count += (current_number == 0)

print(count)      
```
