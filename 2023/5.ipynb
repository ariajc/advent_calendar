```python
import re
```

```python
def construct_map(input:str,map_name:str,vals:list) -> dict:

    values_match = re.search(map_name+r'\n(.+)', input, re.DOTALL)
    values_lines = values_match.group(1).strip().split('\n')

    if '' in values_lines:
        values_lines = values_lines[:values_lines.index('')]
    values_list = [list(map(int, line.split())) for line in values_lines]

    destination = []
    source = []
    for element in values_list:
        destination.extend(list(range(element[0],element[0]+element[2])))
        source.extend(list(range(element[1],element[1]+element[2])))

    res = dict(map(lambda i,j : (i,j) , source,destination))
    
    for element in vals:
        if element not in res:
            res[element] = element
    
    return res

def get_maps_names(input:str)->list:
    map_lines = re.findall(r'.*map.*', input)
    return map_lines

def get_seeds(input:str)->list:
    seeds_match = re.search(r'seeds: (.+)', input)
    seeds_list = list(map(int, seeds_match.group(1).split()))

    return seeds_list
```

```python
def ensemble_dicts(input:str,seeds:list):
    maps = {}
    maps_names = get_maps_names(input)

    vals = seeds

    for element in maps_names:
        print(element)
        map1 = construct_map(input,element,vals)

        if not maps:
            maps = map1
        else:
            for key1, value1 in maps.items():
                # Check if the value of dict1 is a key in dict2
                if value1 in map1:
                    # Combine the key-value pairs
                    maps[key1] = map1[value1]
        
        vals = list(maps.values())

    return maps


    

    
```
